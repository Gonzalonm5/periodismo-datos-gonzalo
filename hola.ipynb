```python
url_ger = "https://api.covid19api.com/country/germany/status/confirmed/live"
df_ger = pd.read_json(url_ger)
```

```python
df_ger
```

```python
df_ger.set_index("Date")["Cases"].plot(title="Casos de covid19 en Alemania")
```

```python
url_kor = "https://api.covid19api.com/country/korea-north/status/confirmed/live"
df_kor = pd.read_json(url_kor)
```

```python
df_kor
```

```python
df_kor.set_index("Date")["Cases"].plot(title="Casos de covid19 en Corea del Norte")
```

```python
casos_esp = df_esp.set_index("Date")["Cases"]
casos_ger = df_ger.set_index("Date")["Cases"]
casos_kor = df_kor.set_index("Date")["Cases"]
```

```python
vs = pd.concat([casos_esp,casos_ger,casos_kor],axis=1)
vs
```

```python
vs.columns = ["España","Alemania","Corea del Norte"]
```

```python
vs
```

```python
vs.plot(title="Casos de covid19 en España, Alemania y Corea del Norte")
```

```python
vs.to_csv("esvsit.csv")
```
